### Insertar los datos del CSV en la tabla creada en la base de datos

```python
import pandas as pd
import mysql.connector
from credenciales import mysql_config


def cargar_csv_en_tabla(df, tabla, conn, batch=1000):
    """Inserta un DataFrame en una tabla de TiDB en lotes de `batch` filas."""
    cursor = conn.cursor()
    cols = ", ".join(df.columns)
    placeholders = ", ".join(["%s"] * len(df.columns))
    sql = f"INSERT INTO {tabla} ({cols}) VALUES ({placeholders})"

    filas = [tuple(None if pd.isna(v) else v for v in row)
             for row in df.itertuples(index=False, name=None)]

    for i in range(0, len(filas), batch):
        cursor.executemany(sql, filas[i:i+batch])
        conn.commit()
        print(f"  insertadas {min(i+batch, len(filas))}/{len(filas)}")
    cursor.close()

cargar_csv_en_tabla(dfmo, "movies", conn, batch=1000)
cargar_csv_en_tabla(dfcr, "credits", conn, batch=1000)
```

### Convierto el realease_date de string a dato tipo fecha para posteriormente poder sacar el año, agrupar por décadas 

```python
df_tablajoin['release_date'] = pd.to_datetime(df_tablajoin['release_date'])
```

### Creo una columna con el año

```python
df_tablajoin['year'] = df_tablajoin['release_date'].dt.year
```

### Creo una columna con la década para poder comparar presupuestos por década

```python
df_tablajoin['decade'] = (df_tablajoin['year'] // 10) * 10
```

### Ver nulos solo en columnas importantes

```python
df_tablajoin[['genres','budget','revenue','release_date','runtime','production_companies']].isna().sum()
```

### Eliminar películas con presupuesto o ingresos 0

```python
df_tablajoin = df_tablajoin[(df_tablajoin['budget'] > 0) & (df_tablajoin['revenue'] > 0)]
```

### Eliminar nulos en runtime (elimino peliculas sin duración)

```python
df_tablajoin = df_tablajoin.dropna(subset=['runtime'])
```

### Filtrar películas demasiado cortas para mantener solo películas con duración mayor a 40 minutos. Evitar cortos o datos raros.

```python
df_tablajoin = df_tablajoin[df_tablajoin['runtime'] > 40]
```

### Elimino nulos en fechas para eliminar películas sin fecha

```python
df_tablajoin = df_tablajoin.dropna(subset=['release_date'])
```

### Eliminar nulos en géneros, elimino filas donde no hay género

```python
df_tablajoin = df_tablajoin.dropna(subset=['genres'])
```

### Compruebo duplicados

```python
df_tablajoin.duplicated().sum()
```

### Creo beneficio. Restar las ganancias (revenue)- el presupuesto (budget)

```python
df_tablajoin['profit'] = df_tablajoin['revenue'] - df_tablajoin['budget']
```

### Creo el ROI restando el beneficio - el presupuesto y dividiendolo entre el prespuesto inicial para calcular la rentabilidad proporcional

```python
df_tablajoin['roi'] = (df_tablajoin['revenue'] - df_tablajoin['budget']) / df_tablajoin['budget']
```

### Como en la columna genres tengo muchos datos, saco solo el nombre para obtener el nombre del genéro de la película

```python
import json
 
df_tablajoin["genres_parsed"] = df_tablajoin["genres"].apply(
    lambda x: [g["name"] for g in json.loads(x)] if pd.notnull(x) else []
)
```

## Fase 7 — EDA 

El EDA es la fase donde **descubrís la historia** del dataset. No es "hacer gráficos sueltos", es **responder preguntas**.

### Estructura recomendada

1. **Descripción general** (5 min): cuántas filas, qué periodo, qué variables.
2. **Univariante** (15 min): distribución de cada variable importante (histogramas, boxplots, value_counts).
3. **Bivariante** (20 min): relación entre variables (scatter, barras por grupo, correlaciones).
4. **Hallazgos** (10 min): 3-5 conclusiones claras con un gráfico cada una.

### Preguntas guía para arrancar (escoged 3-5 según vuestro dataset)

- ¿Cómo evoluciona X a lo largo del tiempo?
- ¿Qué categoría/grupo domina? ¿Hay desigualdad fuerte?
- ¿Hay correlación entre dos variables clave?
- ¿Existen segmentos diferenciados (clusters obvios)?
- ¿Qué porcentaje del total acumulan el top-10 / top-20%? (Pareto)

### Tipos de gráfico por situación

| Lo que quieres mostrar | Gráfico |
|---|---|
| Distribución de una variable numérica | histograma, boxplot |
| Frecuencia de una categórica | barras horizontales (`value_counts().plot.barh()`) |
| Evolución temporal | línea (mejor con `resample('M')` o `groupby(año)`) |
| Relación numérica-numérica | scatter (con `alpha=0.3` si hay muchos puntos) |
| Comparar grupos | boxplot por categoría, barras agrupadas |
| Matriz de correlaciones | heatmap (`sns.heatmap(df.corr())`) |
| Datos geográficos | mapa con `plotly.express.scatter_mapbox` o `folium` |

### Cómo conseguir un buen EDA (no sólo "correcto")

- **Cada gráfico tiene que responder una pregunta concreta**. Si no sabéis qué pregunta responde, sobra.
- **Título descriptivo en cada gráfico**: "Distribución de precios" está mal. "El 80% de los pedidos cuesta menos de 100 €" está bien.
- **Insight escrito al lado de cada gráfico**: una frase explicando qué se ve y qué implica.
- **Storytelling**: que el notebook se lea como una historia (cada apartado lleva al siguiente).


## PREGUNTA 1 - ¿Qué géneros son más rentables?

### Explode para separar los genéros y tener un género por fila

```python
df_generos = df_tablajoin.explode("genres_parsed")
roi_generos = (
    df_generos.groupby("genres_parsed")["roi"]
    .median()
    .sort_values(ascending=False)
    .round(2)
)
roi_generos.head(10)
```

### visualización para responder ¿qué generos son más rentables?

```python
import matplotlib.pyplot as plt
```

```python
roi_generos.count()
```

```python
plt.figure(figsize=(10,5))

roi_generos.plot(kind="bar")

plt.title("Géneros más rentables")
plt.ylabel("ROI mediano")
plt.xlabel("Género")
plt.xticks(rotation=45)

plt.show()

print("El género más rentable es:", roi_generos.index[0])
print("El género menos rentable es:", roi_generos.index[-1])
```

## PREGUNTA 2 - ¿Ha cambiado el presupuesto medio por década?

```python
# Presupuesto medio por década
budget_decade = (
    df_tablajoin.groupby("decade")["budget"]
    .mean()
    .sort_index()
)
```

```python
# Gráfico
plt.figure(figsize=(10,5))

budget_decade.plot(kind="line", marker="o")

plt.title("Presupuesto medio por década")
plt.ylabel("Presupuesto medio")
plt.xlabel("Década")

plt.show()

print("Década con mayor presupuesto:")
print(budget_decade.idxmax())

print("Década con menor presupuesto:")
print(budget_decade.idxmin())
```

## PREGUNTA 5 - ¿La duración de la película influye en el éxito?

```python
df_tablajoin["runtime"].describe()
```

```python
# Crear grupos de duración
df_tablajoin["grupo_duracion"] = pd.cut(
    df_tablajoin["runtime"],
    bins=[0, 90, 120, 150, 300]
)
```

```python
# Recaudación media por duración
runtime_revenue = (
    df_tablajoin.groupby("grupo_duracion")["revenue"]
    .mean()
)
```

```python

# Gráfico
plt.figure(figsize=(8,5))

runtime_revenue.plot(kind="bar")

plt.title("Recaudación según duración")
plt.xlabel("Duración (minutos)")
plt.ylabel("Recaudación media")

plt.xticks(rotation=0)

plt.show()
print("Las películas más largas suelen recaudar más dinero de media.")
```
